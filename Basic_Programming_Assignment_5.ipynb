```python
#1.Write a Python Program to Find LCM?
def gcd(a,b):
    if a==0:
        return b
    return gcd(b%a,a)
a=int(input("Enter the 1st number:"))
b=int(input("Enter the 2nd number:"))
if a>b:
    a,b=b,a
lcm=(a*b)/gcd(a,b)
print("LCM of {} and {} is {}".format(a,b,lcm))
```

```python
#2.	Write a Python Program to Find HCF?
a=int(input("Enter the 1st number:"))
b=int(input("Enter the 2nd number:"))
if a>b:
    a,b=b,a
print("HCF of {} and {} is ".format(a,b),end="")
while a!=0:
    c=a
    a=b%a
    b=c
print("{}".format(b))
```

```python
#3.	Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?
number=int(input("Enter the number"))
print("Number in Binary:",bin(number))
print("Number in Octal:",oct(number))
print("Number in Hexadecimal:",hex(number))
```

```python
#4.	Write a Python Program To Find ASCII value of a character?
char=input("Enter the charcter")
print("ASCII of {} is {}".format(char,ord(char)))
```
