```python
#1.	Write a Python program to check if the given number is a Disarium Number?
def Disarium(number):
    n=number
    sum1=0
    digits=len(str(n))
    while n>0:
        a=n%10
        sum1=sum1+(a**digits)
        digits=digits-1
        n=n//10
    if sum1==number:
        print(number)
```

```python
Disarium(175)
```

```python
#2.	Write a Python program to print all disarium numbers between 1 to 100?
for i in range(1,101):
    Disarium(i)
```

```python
#3.	Write a Python program to check if the given number is Happy Number?
def Happy(num):
        n=num
        sum1=0
        while n>0:
            a=n%10
            sum1+=a**2
            n=n//10
        return sum1
```

```python
#4.	Write a Python program to print all happy numbers between 1 and 100?
for i in range(1,101):
    j=i
    while(j!=1 and j!=4):
        j=Happy(j)
    if j==1:
        print(i)
    
```

```python
#5.	Write a Python program to determine whether the given number is a Harshad Number?
def Harshad(number):
    n=number
    sum1=0
    while n>0:
        a=n%10
        sum1+=a
        n=n//10
    if number%sum1==0:
        print(number)
```

```python
Harshad(25)
```

```python
Harshad(156)
```

```python
#6.	Write a Python program to print all pronic numbers between 1 and 100?
import math
def Pronic(number):
    for i in range(0,int(math.sqrt(number))+1):
        if number==(i*(i+1)):
            return True

for i in range(1,101):
    if(Pronic(i)):
        print(i)
```
