```python
#4.	Write a Python program to swap two variables?
num1=int(input("Enter 1st number"))
num2=int(input("Enter 2nd number"))
print("Before Swapping:-\nnumber1:",num1,"\nnumber2:",num2)
num1,num2=num2,num1
print("After Swapping:-\nnumber1:",num1,"\nnumber2:",num2)
```
