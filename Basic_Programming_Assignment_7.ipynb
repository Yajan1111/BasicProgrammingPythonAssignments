```python
#1.	Write a Python Program to find sum of array?
import array as myarray
inputString=input("Enter the elements of array seprated by spaces:")
list1=inputString.split()
sum=0
for i in range(len(list1)):
    list1[i]=int(list1[i])
arr1=myarray.array('i',list1)
for i in arr1:
    sum+=i
print("Sum of Array is {}".format(sum))
```

```python
#2.	Write a Python Program to find largest element in an array?
largest=arr1[0]
for i in range(1,len(arr1)):
    if largest<arr1[i]:
        largest=arr1[i]
print("Largest number in array is {}".format(largest))
```

```python
#3.	Write a Python Program for array rotation?
#rotating by 2
n=len(arr1)
half=n//2
arr1[:]=arr1[half:n]+arr1[0:half]
arr1
```

```python
#4.	Write a Python Program to Split the array and add the first part to the end?
n=len(arr1)
half=n//2
arr1[:]=arr1[half:n]+arr1[0:half]
arr1
```

```python
#5.	Write a Python Program to check if given array is Monotonic?
#array is [1,2,3,4,5]
for i in range(len(arr1)-1):
    if arr1[i]<=arr1[i+1]:
        count=0
    else:
        for i in range(len(arr1)-1,1,-1):
            if arr1[i]<=arr1[i-1]:
                count=0
            else:
                count=1
                break
if count==0:
    print("Array is monotonic")
else:
    print("Array is not monotonic")
```
