```python
#1.Create a function in python to read the text file and replace specific content of the file.

def replace_content(file_name,word_to_replaced,word_to_add):
    file_reference=open(file_name,"rt")
    #reading the file
    file_data=file_reference.read()
    #replacing the words
    file_data=file_data.replace(word_to_replaced,word_to_add)
    file_reference.close()
    file_reference=open(file_name,"wt")
    #writing the change inside the file
    file_reference.write(file_data)
    file_reference.close()

try:
    filename=input("Enter file name:")
    wordreplace=input("Enter the word to be replaced:")
    wordtoadd=input("enter the word to be added:")
except Exception as e:
    print(e)
else:
    replace_content(filename,wordreplace,wordtoadd)
    print("Word has been replaced!")
```

```python
'''2.Demonstrate use of abstract class, multiple inheritance and decorator in python using examples.
Answer:

Abstract class- Abstract class is a class which has one or more abstract methods in it.Abstract methods has only declaration and no implementation.
When we want that few classes should have same methods with diffrent implementation, we can use abstract class.

Multiple Inheritance:-A child class which inherit from more than 1 base classes it is called as multiple inheritancen 
    
Decorator- Decorators are a way to change, enhance or alter in any way how a function works.Decorators are defined with the @ symbol followed
by the decorator name, just before the function definition.

Example:-'''
from abc import ABC,abstractmethod

class RTO:
    
    def hasIssuedNumberPlate(self):
        print("Number Plate has been issue")

#Vehicle(ABC) makes it abstract class
class Vehicle(ABC):
    
    #decorator used to make it abstract method
    @abstractmethod
    def wheels(self):
        pass
    
#has 2 base class Vehicle & RTO   
class Bike(Vehicle,RTO):
    
    #implementing abstract methods
    def wheels(self):
        print("Has 2 wheels")

splendor=Bike()
splendor.wheels()
splendor.hasIssuedNumberPlate()
```
